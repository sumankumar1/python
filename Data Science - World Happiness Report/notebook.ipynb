```python
d15['year'] = 2015
d16['year'] = 2016
d17['year'] = 2017
d18['year'] = 2018
d19['year'] = 2019
data = pd.concat([d15,d16,d17,d18,d19])
```

```python
data.info()
```

```python
data1 = data.drop(data.columns[[1,4,11,13,14,15,16]],axis=1)
data1.sort_values(by=['Country','year']).head()
```

```python
data1.info()
```

```python
data1.isnull().sum()
```

```python
data2 = data1[data1.isnull().values]
data2
```

```python
data1[data1['Country']=='United Arab Emirates']
```

```python
data2 = data1.fillna(0.25)
```

```python
data2.isnull().sum()
```

```python
score_year_pattern = pd.pivot_table(data2,values='Happiness Score',index=['Country'],columns=['year'])
score_year_pattern =  score_year_pattern.dropna()
score_year_pattern.head()
```

```python
# Sorted in order of Happiness score in 2015
# Observe people in some countries are getting more happier compared to other countries e.g. Denmark, Norway, Finland
score_year_pattern.sort_values(by=2015).plot.barh(figsize=(15,30),stacked=True,rot=15)
```

```python
score_year_pattern.sort_values(by=2015).plot.bar(figsize=(24,5),stacked=True,rot=90)
```

```python
# happiness score increase or decrease by year 
score_year_pattern.sort_values(by=2015).plot.barh(figsize=(15,30),rot=15)
```

```python
top_10_in_2015 = d15.iloc[0:10,:]
top_10_in_2015
```

```python
top10 = data2[data2['Country'].isin(top_10_in_2015['Country'])]
top10 = top10.drop(top10.columns[[1,4]],axis=1).round(decimals=4)
```

```python
top10.sort_values(by='Country').head()
```

```python
i = 0
for country in top10['Country'].unique():
    i = i+1
    ax = top10[top10['Country']==country].iloc[:,:-1].sort_values('Happiness Score').plot.bar(x='Happiness Score',subplots=True,
                                                          figsize=(20,4), layout=(1,5),legend=False, title=country)
    if(i==3):
        break
```

```python
bottom_10_in_2015 = d15.iloc[-5:,:]
bottom_10_in_2015
bottom10 = data2[data2['Country'].isin(bottom_10_in_2015['Country'])]
bottom10 = bottom10.drop(bottom10.columns[[1,4]],axis=1).round(decimals=4)
bottom10.head()
```

```python
i = 0
for country in bottom10['Country'].unique():
    i = i+1
    ax = bottom10[bottom10['Country']==country].iloc[:,:-1].sort_values('Happiness Score').plot.bar(x='Happiness Score',
                                                                                                    subplots=True,figsize=(20,4), layout=(1,5),legend=False, title=country)
    if(i==5):
        break
```
