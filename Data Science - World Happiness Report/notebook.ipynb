## Correlation

```python
cdata = data2.corr(method="pearson")
cdata['Happiness Score']
```

## Overall the world is getting happier
### Health and economy seems to be most important factor 
### Family and freedom are next two important factor
### Next is Trust in government 

```python
 def plot_corr(correlation,location):
    my_colors = ['k', 'k', 'green', 'blue','green','blue','yellow' ,'yellow','red']
    correlation['Happiness Score'].dropna().plot.barh(title='Correlation with Happiness Score',figsize=(10,5),color=my_colors)
    NA = mpatches.Patch(color='blue', label='Very Strong Correlation(>=0.75)')
    EU = mpatches.Patch(color='green', label='Strong Correlation([0.5,0.75])')
    SA = mpatches.Patch(color='yellow', label='Weak Correlation([0.25,<0.5])')
    AP = mpatches.Patch(color='red', label='Very Weak Correlation(<0.25)')
    plt.legend(handles=[NA,EU,SA,AP], loc=location)
plot_corr(cdata,2)   
```

### Top 10 happiest country are getting happier

```python
top10 = data2[(data2['Happiness Rank']<11)]
cdata = top10.corr(method="pearson")
plot_corr(cdata,1) 
```

### Bottom 10 are not getting happier 

```python
bottom10 = data2[ (data2['Happiness Rank']>147)]
cdata = bottom10.corr(method="pearson")
plot_corr(cdata,1) 
```
